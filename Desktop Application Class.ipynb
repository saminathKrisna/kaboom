```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
import matplotlib
matplotlib.use("TkAgg")
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import pandas as pd
from PIL import Image, ImageTk
import os
```

```python
dfa = pd.read_csv("stock_prices(1).csv")
dfb = pd.read_csv("S&P_stock_details.csv")

investing_list = pd.read_csv("reccomendation.csv")["investing_list"].tolist()
        
path = 'C:/Users/USER/College Project/intraQuarter'
stats_path = path + '/_KeyStats'
stock_list = [x[0] for x in os.walk(stats_path)]

ticker_list = []
    
for each_dir in stock_list[1:]:
    each_file = os.listdir(each_dir)
    ticker = each_dir.split('\\')[1]
    ticker_list.append(ticker)

all_companies_list = []

for i in dfb["Symbol"]:
    if i.lower() in ticker_list:
        all_companies_list.append(dfb[dfb["Symbol"] == i]["Name"].tolist()[0])
        
reccomended_stock_list = []

for i in dfb["Symbol"]:
    if i.lower() in investing_list:
        reccomended_stock_list.append(dfb[dfb["Symbol"] == i]["Name"].tolist()[0])

sector_list = []

for i in dfb["Sector"].unique():
    sector_list.append(i)
    
```

```python
class StockPrediction(tk.Tk):
    
    def __init__(self, *args, **kwargs):
        
        tk.Toplevel.__init__(self, *args, **kwargs)
        
        tk.Tk.iconbitmap(self, default = "stock_icon.ico")
        tk.Tk.title(self, "Stock Prediction System")
        
        container = tk.Frame(self)
        container.pack(side = TOP, fill = BOTH, expand = True)
        container.grid_rowconfigure(0, weight = 1)
        container.grid_columnconfigure(0, weight = 1)
    
        self.frames = {}
        
        for f in (StartPage, SectorPage, ResultPage):
            frame = f(container, self)
            self.frames[f] = frame
            frame.grid(row = 0, column = 0, sticky = NSEW)
    
        self.show_frame(StartPage)
    
    def show_frame(self, controller):
        
        frame = self.frames[controller]
        frame.tkraise()
        
```

```python
class SectorPage(tk.Frame):
    
    def __init__(self, parent, controller):
        
        tk.Frame.__init__(self, parent)
        
        left_frame = Frame(self, background = "white")
        left_frame.pack(side = LEFT, fill = Y)

        top_frame = Frame(self, background = "white")
        top_frame.pack(side = TOP, fill = BOTH)

        back_button = Button(left_frame, background = "white", borderwidth = 0)
        photo = Image.open('back_button.png')
        image = ImageTk.PhotoImage(photo)
        back_button.config(image = image)
        back_button.grid(row=0, column=0, padx = (25, 25), pady = (25,25))

        sector_name = tk.Label(top_frame, text = sector_list[2].upper())
        sector_name.config(font=("Courier", 40), bg = "white")
        sector_name.pack(fill = tk.BOTH, ipadx = 50, ipady = 30)

        def sector(selected_sector):
            return dfb[dfb["Sector"] == selected_sector]["Name"].tolist()

        variable = StringVar(self)
        variable.set(sector(sector_list[2])[0])
        drop_down = OptionMenu(self, variable, *sector(sector_list[2]))
        drop_down.config(font=('Courier', 10), bg='white', width = 45)
        drop_down.pack(ipadx = 25, ipady = 15)
```

```python
class ResultPage(tk.Frame):
    
    def __init__(self, parent, controller):
        
        tk.Frame.__init__(self, parent)
        
        left_frame = Frame(self, background = "white")
        left_frame.pack(side = LEFT, fill = Y)

        top_frame = Frame(self, background = "white")
        top_frame.pack(side = TOP, fill = BOTH)

        back_button = Button(left_frame, background = "white", borderwidth = 0)
        photo = Image.open('back_button.png')
        image = ImageTk.PhotoImage(photo)
        back_button.config(image = image)
        back_button.grid(row=0, column=0, padx = (25, 25), pady = (25,25))

        stock_name = tk.Label(top_frame, text = reccomended_stock_list[1].upper() + " (" + investing_list[1].upper() + ")")
        stock_name.config(font=("Courier", 40), bg = "white")
        stock_name.pack(fill = tk.BOTH)

        rec = "Good"

        if stock_name in investing_list:
            rec = "Good"
        else:
            rec = "Bad"

        reccomendation = tk.Label(self, text = "Reccomendation: " + rec)
        reccomendation.config(font=("Courier", 30), bg = "white")
        reccomendation.pack(fill = tk.BOTH)

        f = Figure(figsize = (12,6), dpi = 100)
        a = f.add_subplot(111)
        a.plot(dfa['Date'], dfa[investing_list[1].upper()])
        a.axes.xaxis.set_ticklabels([])

        canvas = FigureCanvasTkAgg(f, self)
        canvas.show()
        canvas.get_tk_widget().pack(fill = tk.BOTH, expand = True)
```
